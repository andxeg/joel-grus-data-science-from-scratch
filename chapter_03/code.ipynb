# Data visualization

# Example

```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt

years = list(range(1950, 2011, 10))
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

# gross domestic product
plt.plot(years, gdp, color='green', marker='o', linestyle='solid')

plt.title("Nominal GDP")
plt.ylabel("bil. $")
plt.show()
```

# Bar chart

```python
# Example 1
movies = ["Энди Холл", "Бен-Гур", "Касабланка", "Ганди", "Вестсайдская история"]
num_oscars = [5, 11, 3, 8, 10]

# bar's width=0.8 by default, add 0.1 to the left coordinats
xs = [4*i+0.1 for i, _ in enumerate(movies)]

# create bars with left X-coords [xs] and heights [num_oscars]
plt.bar(xs, num_oscars)

plt.ylabel("Number of oscars")
plt.xlabel("My favorite films")

plt.xticks([4*i+0.1 for i,_ in enumerate(movies)], movies)
plt.show()
```

```python
# Example 2
from collections import Counter

grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]
decile = lambda grade: grade // 10 * 10
histogram = Counter(decile(grade) for grade in grades)

plt.bar([x for x in histogram.keys()], # shift bar on 4
        histogram.values(),               # bar height
        8)                                # bar width

plt.axis([-5, 105, 0, 5]) # X from -5 to 105, Y from 0 to 5
plt.xticks([10 * i for i in range(11)]) # marsk on x-axis: 0, 10, 20, ..., 100
plt.xlabel("Decile")
plt.ylabel("Number of students")
plt.title("Distribution of marks on the exam")
plt.show()
```

# Bad example

```python
mentions = [500, 505]
years    = [2013, 2014]

plt.bar(years, mentions, 0.8)
plt.xticks(years)
plt.ylabel("Number of reference to data science")
plt.ticklabel_format(useOffset=False)

# NOTE!
# bad practice
plt.axis([2012.5, 2014.5, 499, 506])
plt.title("Huge increase")
plt.show()
```

# How to fix it

```python
mentions = [500, 505]
years    = [2013, 2014]

plt.bar(years, mentions, 0.8)
plt.xticks(years)
plt.ylabel("Number of reference to data science")
plt.ticklabel_format(useOffset=False)

# NOTE!
# Y axis min value - 0
plt.yticks([i for i in range(0, 501, 50)])
plt.axis([2012.5, 2014.5, 0, 550])
plt.title("No so big anymore")
plt.show()
```

# Line charts

```python
variance     = [1, 2, 4, 8, 16, 32, 64, 128, 256] # dispersion
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]

total_error = [x + y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]

plt.plot(xs, variance, 'g-', label='dispersion')      # greed solid line
plt.plot(xs, bias_squared, 'r-.', label='bias^2')     # red dash-dotted line
plt.plot(xs, total_error, 'b:', label='total error')  # blue dotted line

plt.legend(loc=9) # mid-top
plt.xlabel("Model complexity")
plt.title("Trade-off between bias and variance")
plt.show()
```

# Scatter charts

```python
friends = [70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
# [chr(i) for i in range(ord('a'), ord('j'))]
labels  = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(friends, minutes)

for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label,
                 xy=(friend_count, minute_count),
                 xytext=(5, -5),
                 textcoords='offset points')

plt.title("The relationship between number of friends and number of minutes")
plt.xlabel("Number of friends")
plt.ylabel("Number of minutes")
plt.show()
```

# Bad example

```python
# NOTE!
# You can get misrepresented chart
# if limit of axis matplotlib defines automatically
test_1_grades = [99, 90, 85, 97, 80]
test_2_grades = [100, 85,60, 90, 70]

plt.scatter(test_1_grades, test_2_grades)
plt.title("Incomparable axis")
plt.xlabel("Marks for test #1")
plt.ylabel("Marks for test #2")
plt.show()
```

# How to fix it

```python
# Add plt.axis("equal")

test_1_grades = [99, 90, 85, 97, 80]
test_2_grades = [100, 85,60, 90, 70]

plt.scatter(test_1_grades, test_2_grades)
plt.title("Incomparable axis")
plt.xlabel("Marks for test #1")
plt.ylabel("Marks for test #2")
plt.axis("equal")
plt.show()
```
